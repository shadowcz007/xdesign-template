```python
import requests
```

```python
res = requests.get('https://api.github.com/events')
```

```python
help(res)
```

```python
res=res.json()
```

```python
for r in res:
#     print(r.keys())
#     print()
    
    for key in r.keys():
        print('键名',key)
        print('键值',r[key])
        print('------------')
    print()
```
